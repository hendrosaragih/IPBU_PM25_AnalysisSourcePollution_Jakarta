```python
import cdsapi
import os
import glob
from pathlib import Path

class ERA5DownloaderFixed:
    def __init__(self):
        self.client = cdsapi.Client()
    
    def download_era5_data(self, year, month, variables=None):
        """Download data ERA5 dengan parameter yang benar"""
        if variables is None:
            variables = [
                '10m_u_component_of_wind', 
                '10m_v_component_of_wind',
                '2m_dewpoint_temperature', 
                '2m_temperature',
                'total_precipitation',
                'surface_pressure',
                'mean_sea_level_pressure'
            ]
        
        # Area untuk Jakarta dan sekitarnya
        area = [-6.4, 106.6, -6.1, 107.1]  [North, West, South, East]
        
        request = {
            'product_type': 'reanalysis',
            'format': 'netcdf',
            'variable': variables,
            'year': year,
            'month': month,
            'day': [f'{i:02d}' for i in range(1, 32)],
            'time': [f'{i:02d}:00' for i in range(24)],
            'area': area,
        }
        
        output_dir = 'C:\\Users\\user\\OneDrive\\IPB\\Thesis\\02. Development\\01. Data Praprocessing\\raw_data_fixed'
        os.makedirs(output_dir, exist_ok=True)
        
        output_file = os.path.join(output_dir, f'era5_hourly_{year}_{month}.nc')
        
        print(f"üì• Downloading {year}-{month}...")
        
        try:
            self.client.retrieve('reanalysis-era5-single-levels', request, output_file)
            
            # Check file size
            file_size = os.path.getsize(output_file) / (1024*1024)  # MB
            print(f"‚úÖ Download selesai: {output_file} ({file_size:.1f} MB)")
            
            return output_file
            
        except Exception as e:
            print(f"‚ùå Gagal download {year}-{month}: {str(e)}")
            return None

# Download data untuk testing
downloader = ERA5DownloaderFixed()
downloader.download_era5_data('2022', '01')
```
