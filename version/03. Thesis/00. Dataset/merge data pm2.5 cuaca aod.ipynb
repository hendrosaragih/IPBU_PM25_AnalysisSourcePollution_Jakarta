```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def merge_data(path_pm25, path_weather, path_aod):
    df_pm25 = pd.read_csv(path_pm25)
    df_pm25['datetime'] = pd.to_datetime(df_pm25['datetime']).dt.date

    
    df_pm25.replace('-', np.nan, inplace=True)
    df_pm25.replace(0, np.nan, inplace=True)
    
    df_pm25['ISPU PM2.5'] = pd.to_numeric(df_pm25['ISPU PM2.5'], errors='coerce')
    
    df_pm25_daily = df_pm25.groupby('datetime', as_index=False)['ISPU PM2.5'].mean()
    df_pm25_daily = df_pm25_daily[["ISPU PM2.5", "datetime"]]

    df_weather = pd.read_csv(path_weather)

    df_pm25_daily['datetime'] = pd.to_datetime(df_pm25_daily['datetime']).dt.date
    df_weather['datetime'] = pd.to_datetime(df_weather['datetime'], format='%m/%d/%Y').dt.date

    columns_needed = ['datetime', 'temp', 'dew', 'humidity', 'precip', 'windspeed']
    df_weather = df_weather[columns_needed]

    df_merged = pd.merge(df_pm25_daily, df_weather, on='datetime', how='inner')

    df_aod = pd.read_csv(path_aod)
    df_aod['datetime'] = pd.to_datetime(df_aod['datetime'], format='%Y-%m-%d').dt.date
    columns_needed = ['datetime','AOD']
    df_aod = df_aod[columns_needed]

    df_merged = pd.merge(df_merged, df_aod, on='datetime', how='left')

    return df_merged




    
```

```python
path_pm25 = "C:\\AData\\kuliah\\sem 7\\Tugas Akhir\\Kode Program\\merged data pm2.5\\merged_data_Bundaran HI.csv"
path_weather = "C:\\AData\\kuliah\\sem 7\\Tugas Akhir\\Kode Program\\data cuaca\\bundaran_hi 2022-01-01 to 2024-12-31.csv"
path_aod = "C:\\AData\\kuliah\\sem 7\\Tugas Akhir\\Kode Program\\data aod himawari\\data csv\\bundaran_hi.csv"

df = merge_data(path_pm25, path_weather, path_aod)
```
