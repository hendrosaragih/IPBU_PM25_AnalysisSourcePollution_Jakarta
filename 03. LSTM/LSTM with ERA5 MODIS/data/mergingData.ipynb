```python
import pandas as pd
import glob
import os

# Path folder tempat semua file CSV berada
folder_path = r"C:\Users\user\OneDrive\IPB\Thesis\02. Development\03. LSTM\LSTM with ERA5 MODIS\data"

# List file CSV yang ingin digabungkan
csv_files = [
    "bundaran_hi.csv",
    "kelapa_gading.csv",
    "kebun_jeruk.csv",
    "jagakarsa.csv",
    "lubang_buaya.csv"
]

# List untuk menampung data
df_list = []

for file in csv_files:
    full_path = os.path.join(folder_path, file)

    # Baca CSV
    df = pd.read_csv(full_path)

    # Tambahkan kolom station_name dari nama file
    station_name = file.replace(".csv", "")
    df["station_name"] = station_name

    # Simpan ke list
    df_list.append(df)

# Gabungkan semua dataframe
final_df = pd.concat(df_list, ignore_index=True)

# Simpan ke satu file CSV
output_path = os.path.join(folder_path, "pm25_aod_meteorology_merged.csv")
final_df.to_csv(output_path, index=False)

print("Selesai! File gabungan disimpan sebagai:", output_path)
```
